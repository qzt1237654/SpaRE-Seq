```python
import os

def print_directory_contents(path):
    for root, dirs, files in os.walk(path):
        # 排除掉一些没用的隐藏文件夹，比如 git 和 缓存
        dirs[:] = [d for d in dirs if not d.startswith('.') and d != '__pycache__' and 'egg-info' not in d]
        
        level = root.replace(path, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root) if os.path.basename(root) else path))
        
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            if not f.startswith('.'):
                print('{}{}'.format(subindent, f))

# 运行这一行
print_directory_contents(r'C:\Users\28616\Desktop\SpaRE-Seq')
```
