```python
import pandas as pd
import numpy as np
import os

gene_dir = r"C:\Users\28616\Desktop\gene_cluster_expression"
processed_gene_dir = r"C:\Users\28616\Desktop\gene_processed"
os.makedirs(processed_gene_dir, exist_ok=True)

sample_ids = ["49", "50", "51", "52", "53", "54"]

for s_id in sample_ids:
    file_path = os.path.join(gene_dir, f"Sample_{s_id}_gene_matrix.csv")
    if not os.path.exists(file_path): continue
    
    # Read matrix (Rows: Genes, Cols: Cluster_0-8)
    df_gene = pd.read_csv(file_path, index_col=0)
    
    # 1. Normalization: log2(counts + 1)
    df_log = np.log2(df_gene + 1)
    
    # 2. Transpose (Now Rows: Cluster_0-8, Cols: Genes)
    df_processed = df_log.T
    
    # Save processed matrix
    df_processed.to_csv(os.path.join(processed_gene_dir, f"Sample_{s_id}_gene_processed.csv"))

print("Step 1: Gene normalization and transposition complete.")
```
