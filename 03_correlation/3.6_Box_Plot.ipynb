```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

input_dir = r"C:\Users\28616\Desktop\Correlation_Results"
output_dir = r"C:\Users\28616\Desktop\BoxPlots_Correlation\Pearson_R"
os.makedirs(output_dir, exist_ok=True)

sample_ids = ["49", "50", "51", "52", "53", "54"]

# 选择你想要的指标：'Spearman_Rho' 或 'Pearson_R'
target_metric = 'Pearson_R' 

for s_id in sample_ids:
    file_path = os.path.join(input_dir, f"Corr_Results_Sample_{s_id}.csv")
    if not os.path.exists(file_path):
        continue
    
    df = pd.read_csv(file_path)
    
    plt.figure(figsize=(10, 6))
    
    # 画箱形图
    # x 是基因名，y 是你选的相关系数
    sns.boxplot(data=df, x='Gene', y=target_metric, palette='Set3', showfliers=False)
    
    # 叠加一个“小提琴图”或者“条带图”可以让图看起来更高大上（可选）
    sns.stripplot(data=df, x='Gene', y=target_metric, color='black', size=1, alpha=0.1)

    plt.axhline(0, color='red', linestyle='--', linewidth=1) # 增加一条 0 基准线
    plt.title(f"Sample {s_id}: Distribution of {target_metric} across 3 Genes")
    plt.ylabel(f"{target_metric} Value")
    plt.xlabel("Target Genes")
    
    # 保存
    plt.savefig(os.path.join(output_dir, f"Sample_{s_id}_Boxplot_person.png"), dpi=300, bbox_inches='tight')
    plt.close()

print(f"箱形图已生成至: {output_dir}")
```
