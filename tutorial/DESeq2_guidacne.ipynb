```python
import os
import spare_seq.DESeq2 as dp

# 1. Configuration
data_raw_dir = r"path/to/your/ATOI_h5ad_folder"
list_dir     = r"path/to/your/clustering_txt_files"
out_dir      = r"path/to/output_directory"
gtf_path     = r"path/to/your/genome_annotation.gtf"
db_path      = r"path/to/save/gtf_database.db"


# 2. Run Scenario A: Stress vs Control (includes auto Volcano plotting)
stress_out = os.path.join(out_dir, "stress_vs_control")

dp.run_test_vs_control(
    data_raw_dir=data_raw_dir,
    list_dir=list_dir,
    output_dir=stress_out,
    ctrl_samples=['50', '52', '51'],   # control group sample ID
    stress_samples=['54', '49', '53'], # test group sample ID
    n_clusters=6,
    layer_key='AGcount_A',
    p_thresh=0.05,
    lfc_thresh=1.0,      # |Log2FC| threshold
    top_n_labels=5,      # The volcano map automatically marks the five most prominent sites
    plot_barplot=True    
)

# 3. Run Scenario B: Target Cluster vs Others (includes auto Volcano plotting)
target_out = os.path.join(out_dir, "cluster_vs_others")
dp.run_cluster_vs_others(
    data_raw_dir=data_raw_dir,
    list_dir=list_dir,
    output_dir=target_out,
    samples=['50', '52', '51', '54', '49', '53'],
    n_clusters=6,
    layer_key='AGcount_A'
    p_thresh=0.05,
    lfc_thresh=1.0,
    top_n_labels=5, 
    plot_barplot=True  # Auto-labels top 5 genes on Volcano plots
)

# 4. Run Genomic Annotation on the generated significant sites
dp.run_annotation(
    results_dir=os.path.join(target_out, "significant_results"),
    gtf_path=gtf_path,
    db_path=db_path,
    num_clusters=6
)
```
