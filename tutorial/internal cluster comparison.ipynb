```python
import spare_seq.preprocessing as prep
import spare_seq.differential as diff

# 1. Configuration (Line 3: Control Only)
CTRL_SAMPLES = ["50", "51", "52"]
RAW_ATOI_DIR = "PATH/TO/YOUR/ATOI_H5AD_DIR"
CLUSTER_LIST_DIR = "PATH/TO/YOUR/CLUSTER_TXT_DIR"
OUTPUT_DIR = "./OnlyC_Analysis_Results"

# 2. Extract Data and Apply Two-Step Filtering
# Creates Target Cluster and Except_Cluster matrices
prep.prepare_onlyC_cluster_vs_others_matrices(
    data_raw_dir=RAW_ATOI_DIR,
    list_dir=CLUSTER_LIST_DIR,
    output_dir=f"{OUTPUT_DIR}/split_matrices",
    samples=CTRL_SAMPLES
)

# 3. Execute Internal Comparison in DESeq2
# Uses model: ~ Type + Group + Type:Group
diff.run_deseq2_onlyC_cluster_vs_others(
    input_dir=f"{OUTPUT_DIR}/split_matrices",
    output_dir=f"{OUTPUT_DIR}/deseq2_raw"
)

# 4. Clean formatting and Extract Significant Sites
# Retains sites with P-value < 0.05 and |log2FC| > 1.0
diff.filter_onlyC_deseq2_results(
    deseq2_dir=f"{OUTPUT_DIR}/deseq2_raw",
    output_dir=f"{OUTPUT_DIR}/filtered_significance",
    p_thresh=0.05,
    lfc_thresh=1.0
)
```
