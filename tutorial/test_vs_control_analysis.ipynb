```python
import spare_seq.differential_pipeline as dp
import os

# ==========================================
# 1. Configuration (Paths & Samples)
# ==========================================
# Directory containing raw A-to-I mutation data (*_AtoI_tissue.h5ad)
data_raw_dir = r"PATH/TO/RAW_ATOI_DATA" 
# Directory containing cluster lists (.txt) generated by cluster.py
list_dir = r"./clustering_output/cluster_txt_files" 
# Main output directory for DESeq2 results
output_dir = r"./differential_output"

ctrl_samples = ['50', '52', '51']
stress_samples = ['54', '49', '53']
all_samples = ctrl_samples + stress_samples

# ==========================================
# 2. Scenario A: Stress vs. Control
# ==========================================
# Find significantly differentially edited sites between Stress and Control within the same Cluster
print("🚀 Starting Stress vs Control Pipeline...")
dp.run_pipeline_stress_vs_control(
    data_raw_dir=data_raw_dir,
    list_dir=list_dir,
    output_dir=os.path.join(output_dir, "stress_vs_control"),
    ctrl_samples=ctrl_samples,
    stress_samples=stress_samples,
    n_clusters=6,
    p_thresh=0.05,       # Significance P-value threshold
    lfc_thresh=1.0       # Log2FoldChange threshold
)

# ==========================================
# 3. Scenario B: Target Cluster vs. Others
# ==========================================
# Find marker editing sites specific to a Target Cluster (compared to all other clusters)
print("🚀 Starting Cluster vs Others Pipeline...")
dp.run_pipeline_cluster_vs_others(
    data_raw_dir=data_raw_dir,
    list_dir=list_dir,
    output_dir=os.path.join(output_dir, "cluster_vs_others"),
    samples=all_samples,
    n_clusters=6,
    p_thresh=0.05,
    lfc_thresh=1.0
)

print("✅ All differential analyses completed! Check your 'significant_results' folders.")
```
