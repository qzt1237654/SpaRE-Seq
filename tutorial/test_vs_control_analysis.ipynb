```python
import spare_seq.preprocessing as prep
import spare_seq.differential as diff

# 1. Parameter Configuration
CTRL_SAMPLES = ["50", "51", "52"]
STRESS_SAMPLES = ["49", "53", "54"]
RAW_ATOI_DIR = "PATH/TO/YOUR/ATOI_H5AD_DIR"
CLUSTER_LIST_DIR = "PATH/TO/YOUR/CLUSTER_TXT_DIR"
OUTPUT_DIR = "./stress_vs_control_results"

# 2. Generate Matrices (Cross-Group Mode)
# Workflow: Data extraction -> Site alignment -> Median depth filter -> 4-way matrix split
prep.prepare_stress_vs_control_matrices(
    data_raw_dir=RAW_ATOI_DIR,
    list_dir=CLUSTER_LIST_DIR,
    output_dir=f"{OUTPUT_DIR}/split_matrices",
    ctrl_samples=CTRL_SAMPLES,
    stress_samples=STRESS_SAMPLES
)

# 3. Run Differential Analysis (Interaction Model: ~Type + Trt + Type:Trt)
diff.run_deseq2_stress_vs_control(
    split_dir=f"{OUTPUT_DIR}/split_matrices",
    output_dir=f"{OUTPUT_DIR}/deseq2_raw"
)
```
