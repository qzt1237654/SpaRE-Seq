```python
import os
import spare_seq.cluster as clu

# 1. Initialize R and spCLUE environment
spCLUE = clu.init_spclue_env(
    r_home=r"D:\R-4.4.3", 
    r_user=r"C:\Users\28616\anaconda3\envs\spCLUE\lib\site-packages\rpy2", 
    spclue_path=r"C:\Users\28616\Downloads\spCLUE-main"
)

# 2. Set parameters
samples = ['50', '52', '51', '54', '49', '53']
h5ad_temp = r"C:\Users\28616\Desktop\spatialDE_rawdata\GSM81924{name}_GeneID_tissue.h5ad"
out_dir = r"C:\Users\28616\Desktop\SpaRE_Seq_Full_Test\01_Clustering"
os.makedirs(out_dir, exist_ok=True)

# 3. Run clustering pipeline
adata = clu.run_spclue_clustering(
    sample_names=samples, 
    h5ad_template_path=h5ad_temp, 
    spclue_mod=spCLUE, 
    n_clusters=6
)

# 4. Generate QC and Spatial plots
clu.plot_umap_qc(adata, save_path=os.path.join(out_dir, "UMAP_QC.png"))
clu.plot_spatial_clusters(
    adata, sample_names=samples, 
    save_path=os.path.join(out_dir, "Spatial_Clusters.png"),
    flip_dict={"50": "lr", "52": "lr", "54": "lr", "49": "ud"}
)

# 5. Export clean bin IDs for downstream analysis
clu.export_clean_cluster_bins(
    adata, sample_names=samples, 
    h5ad_template_path=h5ad_temp, 
    output_dir=os.path.join(out_dir, "cluster_txt_files"), 
    n_clusters=6
)
```
