```python
import pandas as pd
import os

input_dir = r"C:\Users\28616\Desktop\Filtered_Matrices"  # 假设这是你处理后的矩阵目录
output_dir = r"C:\Users\28616\Desktop\Split_Matrices"
if not os.path.exists(output_dir): os.makedirs(output_dir)

# 定义样本分组
control_samples = ["50", "51", "52"]
stress_samples = ["49", "53", "54"]

print("✂️ 开始执行矩阵转置与分组拆分...")

for c in range(9):
    file_path = os.path.join(input_dir, f"Cluster_{c}_new.csv")
    
    if os.path.exists(file_path):
        # 1. 读取并转置 (使得位点在行，样本列在列)
        df = pd.read_csv(file_path, index_col=0).T
        
        # 2. 构建列名索引列表
        # Control组列名
        ctrl_cols_G = [f"{s}_G" for s in control_samples]
        ctrl_cols_A = [f"{s}_A" for s in control_samples]
        # Stress组列名
        stress_cols_G = [f"{s}_G" for s in stress_samples]
        stress_cols_A = [f"{s}_A" for s in stress_samples]
        
        # 3. 提取子矩阵
        # 确保只提取列名存在于 df 中的列
        df_G_ctrl = df[[col for col in ctrl_cols_G if col in df.columns]]
        df_A_ctrl = df[[col for col in ctrl_cols_A if col in df.columns]]
        df_G_stress = df[[col for col in stress_cols_G if col in df.columns]]
        df_A_stress = df[[col for col in stress_cols_A if col in stress_cols_A if col in df.columns]]
        
        # 4. 保存四部分
        df_G_ctrl.to_csv(os.path.join(output_dir, f"G_control_cluster{c}.csv"))
        df_A_ctrl.to_csv(os.path.join(output_dir, f"A_control_cluster{c}.csv"))
        df_G_stress.to_csv(os.path.join(output_dir, f"G_stress_cluster{c}.csv"))
        df_A_stress.to_csv(os.path.join(output_dir, f"A_stress_cluster{c}.csv"))
        
        print(f"✅ Cluster {c}: 已拆分为 4 个部分。")

print("\n🎉 所有矩阵拆分任务完成！")
```
