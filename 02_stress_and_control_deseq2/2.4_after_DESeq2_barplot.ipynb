```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. 路径配置 ---
input_dir = r"C:\Users\28616\Desktop\Filtered_Results"
output_plot_dir = r"C:\Users\28616\Desktop\Plots" # 图片保存路径

if not os.path.exists(output_plot_dir):
    os.makedirs(output_plot_dir)

# --- 2. 收集数据 ---
data = []

print("📊 正在统计各个 Cluster 的显著位点数量...")
for c in range(6):
    file_path = os.path.join(input_dir, f"Cluster{c}_Significant_Sites.csv")
    
    if os.path.exists(file_path):
        df = pd.read_csv(file_path, index_col=0)
        
        # 统计上调 (log2FC > 1) 和下调 (log2FC < -1) 的个数
        up_count = len(df[df['log2FoldChange'] > 1])
        down_count = len(df[df['log2FoldChange'] < -1])
        
        data.append({'Cluster': f'Cluster {c}', 'Count': up_count, 'Direction': 'Up-regulated (Stress > Ctrl)'})
        data.append({'Cluster': f'Cluster {c}', 'Count': down_count, 'Direction': 'Down-regulated (Stress < Ctrl)'})
    else:
        # 如果某个 Cluster 没结果，记为 0
        data.append({'Cluster': f'Cluster {c}', 'Count': 0, 'Direction': 'Up-regulated (Stress > Ctrl)'})
        data.append({'Cluster': f'Cluster {c}', 'Count': 0, 'Direction': 'Down-regulated (Stress < Ctrl)'})

df_plot = pd.DataFrame(data)

# --- 3. 开始绘图 ---
plt.figure(figsize=(10, 6))

# 使用 seaborn 画水平柱状图
# 设置调色板，红色代表上调，蓝色代表下调
colors = {'Up-regulated (Stress > Ctrl)': '#e74c3c', 'Down-regulated (Stress < Ctrl)': '#3498db'}

ax = sns.barplot(
    data=df_plot, 
    y='Cluster', 
    x='Count', 
    hue='Direction', 
    palette=colors
)

# --- 4. 图表修饰 ---
plt.title('Number of Stress-Responsive A-to-I Editing Sites per Cluster', fontsize=15, pad=15)
plt.xlabel('Number of Significant Sites (p < 0.05, |log2FC| > 1)', fontsize=12)
plt.ylabel('Brain Region / Cell Type', fontsize=12)

# 在柱子旁边加上具体的数字，显得更专业
for p in ax.patches:
    width = p.get_width()
    if width > 0: # 只有大于0才标数字
        ax.text(width + max(df_plot['Count']) * 0.01,  # X轴位置稍微往右偏移一点
                p.get_y() + p.get_height() / 2,        # Y轴居中
                f'{int(width)}',                       # 文本内容
                va='center', fontsize=10)

plt.legend(title='Editing Change under Stress', loc='lower right')
plt.grid(axis='x', linestyle='--', alpha=0.6) # 加点网格线方便看刻度

# --- 5. 保存图片 ---
save_path = os.path.join(output_plot_dir, "Stress_Responsive_Sites_Barplot.png")
plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.close()

print(f"✨ 绘图完成！图片已保存至: {save_path}")
```
