```python
import pandas as pd
import os

# --- 1. 路径配置 ---
res_base_dir = r"C:\Users\28616\Desktop\DESeq2_Results"
# 筛选后新 CSV 的存放路径
output_dir = r"C:\Users\28616\Desktop\Filtered_Results"

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

print("🚀 开始筛选显著位点...")

for c in range(6):
    # 根据你提供的路径：注意文件名大小写需与实际一致
    res_path = os.path.join(res_base_dir, f"DESeq2_Result_cluster{c}.csv")
    
    if os.path.exists(res_path):
        # 读取数据，设置第一列（位点ID）为索引
        df = pd.read_csv(res_path, index_col=0)
        
        # 1. 核心筛选逻辑：p < 0.05 且 |log2FC| > 1
        # 同时利用 dropna 剔除你之前提到的 NA 行
        df_sig = df[
            (df['pvalue'] < 0.05) & 
            ((df['log2FoldChange'] > 1) | (df['log2FoldChange'] < -1))
        ].copy()
        
        # 2. 重构数据框：只保留 pvalue 和 log2FoldChange
        # .copy() 是为了避免修改原始数据时的警告
        df_final = df_sig[['pvalue', 'log2FoldChange']]
        
        # 3. 排序：按照 log2FoldChange 从大到小排序
        df_final = df_final.sort_values(by='log2FoldChange', ascending=False)
        
        # 4. 保存为新的 CSV 文件
        output_file = os.path.join(output_dir, f"Cluster{c}_Significant_Sites.csv")
        df_final.to_csv(output_file)
        
        # 5. 同时保存一个纯位点的 txt list (方便后续其他分析调用)
        list_file = os.path.join(output_dir, f"Cluster{c}_sites_list.txt")
        with open(list_file, 'w') as f:
            for site in df_final.index:
                f.write(f"{site}\n")
        
        print(f"✅ Cluster {c}: 筛选出 {len(df_final)} 个位点 -> 已存至 {os.path.basename(output_file)}")
    else:
        print(f"⚠️ 找不到文件: {res_path}")

print(f"\n✨ 筛选完成！请在 {output_dir} 查看结果。")
```
